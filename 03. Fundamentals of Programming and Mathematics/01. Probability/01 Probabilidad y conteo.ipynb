## 1.2 Espacios muestrales y el Mundo de las Piedras (Pebble World)

El marco matemático de la probabilidad se construye en torno a conjuntos. Imagina que se realiza un experimento cuyo resultado es uno entre un conjunto de posibles resultados. Antes de que se lleve a cabo el experimento, se desconoce cuál será el resultado; después, el resultado «cristaliza» en el resultado real.

**Definición 1.2.1 (Espacio muestral y evento).** El espacio muestral $S$ de un experimento es el conjunto de todos los posibles resultados del experimento. Un evento $A$ es un subconjunto del espacio muestral $S$, y decimos que $A$ ocurrió si el resultado real está en $A$.

La teoría de conjuntos es muy útil en probabilidad, ya que proporciona un lenguaje rico para expresar y trabajar con eventos. Las operaciones de conjuntos, especialmente uniones, intersecciones y complementos, facilitan la construcción de nuevos eventos en términos de eventos ya definidos.

Los eventos pueden describirse en español o en notación de conjuntos. A veces la descripción en español es más fácil de interpretar, mientras que la notación de conjuntos es más fácil de manipular.

## 1.3 Definición ingenua de probabilidad

Históricamente, la definición más temprana de la probabilidad de un evento consistía en contar el número de formas en que el evento podría ocurrir y dividir por el número total de resultados posibles del experimento. Llamamos a esto la definición ingenua ya que es restrictiva y se basa en suposiciones fuertes; sin embargo, es importante de entender y útil cuando no se usa incorrectamente.

**Definición 1.3.1 (Definición ingenua de probabilidad).** Sea $A$ un evento para un experimento con un espacio muestral finito $S$. La probabilidad ingenua de $A$ es:

$$P_{\text{naive}}(A) = \frac{|A|}{|S|} = \frac{\text{número de resultados favorables a } A}{\text{número total de resultados en } S}$$

(Usamos $|A|$ para denotar el tamaño de $A$)

En términos del Mundo de las Piedras, la definición ingenua simplemente dice que la probabilidad de $A$ es la fracción de piedras que están en $A$.

En general:

$$P_{\text{naive}}(A^c) = \frac{|A^c|}{|S|} = \frac{|S| - |A|}{|S|} = 1 - \frac{|A|}{|S|} = 1 - P_{\text{naive}}(A)$$

Una buena estrategia cuando se intenta encontrar la probabilidad de un evento es comenzar pensando si será más fácil encontrar la probabilidad del evento o la probabilidad de su complemento. Las leyes de De Morgan son especialmente útiles en este contexto, ya que puede ser más fácil trabajar con una intersección que con una unión, o viceversa.

La definición ingenua es muy restrictiva ya que requiere que $S$ sea finito, con masa igual para cada piedra. Pero hay varios tipos importantes de problemas donde la definición ingenua es aplicable:

- Cuando hay simetría en el problema que hace que los resultados sean igualmente probables
- Cuando los resultados son igualmente probables por diseño
- Cuando la definición ingenua sirve como un modelo nulo útil

## 1.4 Cómo contar

Calcular la probabilidad ingenua de un evento $A$ implica contar el número de piedras en $A$ y el número de piedras en el espacio muestral $S$. A menudo, los conjuntos que necesitamos contar son extremadamente grandes.

### 1.4.1 Regla de la multiplicación

En algunos problemas, podemos contar directamente el número de posibilidades usando un principio básico pero versátil llamado la regla de la multiplicación. Veremos que la regla de la multiplicación conduce naturalmente a reglas de conteo para muestreo con reemplazo y muestreo sin reemplazo, dos escenarios que surgen con frecuencia en probabilidad y estadística.

**Teorema 1.4.1 (Regla de la multiplicación).** Considera un experimento compuesto que consiste en dos sub-experimentos, Experimento A y Experimento B. Supón que el Experimento A tiene $a$ posibles resultados, y para cada uno de esos resultados el Experimento B tiene $b$ posibles resultados. Entonces el experimento compuesto tiene $ab$ posibles resultados.

**Cuidado 1.4.2:** A menudo es más fácil pensar en los experimentos como si estuvieran en orden cronológico, pero no hay ningún requisito en la regla de la multiplicación de que el Experimento A tenga que realizarse antes que el Experimento B.

**Ejemplo 1.4.3 (Corredores).** Supón que 10 personas están corriendo una carrera. Asume que los empates no son posibles y que los 10 completarán la carrera, por lo que habrá ganadores bien definidos de primer, segundo y tercer lugar. ¿Cuántas posibilidades hay para los ganadores de primer, segundo y tercer lugar?

**Solución:** Hay 10 posibilidades para quién obtiene el primer lugar, luego una vez que eso está fijado hay 9 posibilidades para quién obtiene el segundo lugar, y una vez que ambos están fijados hay 8 posibilidades para el tercer lugar. Entonces, por la regla de la multiplicación, hay $10 \times 9 \times 8 = 720$ posibilidades.

**Subconjuntos.** Un conjunto con $n$ elementos tiene $2^n$ subconjuntos, incluyendo el conjunto vacío $\emptyset$ y el conjunto mismo. Esto se deduce de la regla de la multiplicación ya que para cada elemento, podemos elegir si incluirlo o excluirlo.

Podemos usar la regla de la multiplicación para llegar a fórmulas para muestreo con y sin reemplazo. Muchos experimentos en probabilidad y estadística pueden interpretarse en uno de estos dos contextos, por lo que es atractivo que ambas fórmulas se deriven directamente del mismo principio básico de conteo.

**Teorema 1.4.7 (Muestreo con reemplazo).** Considera $n$ objetos y hacer $k$ elecciones de ellos, una a la vez con reemplazo (es decir, elegir un cierto objeto no impide que sea elegido nuevamente). Entonces hay $n^k$ posibles resultados (donde el orden importa, en el sentido de que, por ejemplo, elegir el objeto 3 y luego el objeto 7 se cuenta como un resultado diferente que elegir el objeto 7 y luego el objeto 3).

**Teorema 1.4.8 (Muestreo sin reemplazo).** Considera $n$ objetos y hacer $k$ elecciones de ellos, una a la vez sin reemplazo (es decir, elegir un cierto objeto impide que sea elegido nuevamente). Entonces hay $n(n-1)\cdots(n-k+1)$ posibles resultados para $1 \leq k \leq n$, y 0 posibilidades para $k > n$ (donde el orden importa). Por convención, $n(n-1)\cdots(n-k+1) = n$ para $k = 1$.

**Ejemplo 1.4.9 (Permutaciones y factoriales).** Una permutación de $1, 2, \ldots, n$ es un arreglo de ellos en algún orden, por ejemplo, $3, 5, 1, 2, 4$ es una permutación de $1, 2, 3, 4, 5$. Por el Teorema 1.4.8 con $k = n$, hay $n!$ permutaciones de $1, 2, \ldots, n$. Por ejemplo, hay $n!$ formas en que $n$ personas pueden hacer fila para comprar helado. (Recuerda que $n!$ es $n(n-1)(n-2)\cdots 1$ para cualquier entero positivo $n$, y $0! = 1$.)

Los Teoremas 1.4.7 y 1.4.8 son teoremas sobre conteo, pero cuando la definición ingenua aplica, podemos usarlos para calcular probabilidades. Esto nos lleva a nuestro siguiente ejemplo, un problema famoso en probabilidad llamado el problema del cumpleaños. La solución incorpora tanto muestreo con reemplazo como muestreo sin reemplazo.

**Ejemplo 1.4.10 (Problema del cumpleaños).** Hay $k$ personas en una habitación. Asume que el cumpleaños de cada persona tiene la misma probabilidad de ser cualquiera de los 365 días del año (excluimos el 29 de febrero), y que los cumpleaños de las personas son independientes. ¿Cuál es la probabilidad de que al menos un par de personas en el grupo tengan el mismo cumpleaños?

**Solución:**

Hay $365^k$ formas de asignar cumpleaños a las personas en la habitación, ya que podemos imaginar que los 365 días del año se muestrean $k$ veces, con reemplazo. Por suposición, todas estas posibilidades son igualmente probables, por lo que la definición ingenua de probabilidad aplica.

Usada directamente, la definición ingenua dice que solo necesitamos contar el número de formas de asignar cumpleaños a $k$ personas de tal manera que haya dos personas que compartan un cumpleaños.

Pero este problema de conteo es difícil, ya que podría ser Emma y Steve quienes comparten un cumpleaños, o Steve y Naomi, o los tres, o los tres podrían compartir un cumpleaños mientras otros dos en el grupo comparten un cumpleaños diferente, o varias otras posibilidades.

En su lugar, contemos el complemento: el número de formas de asignar cumpleaños a $k$ personas de tal manera que ninguna dos personas compartan un cumpleaños. Esto equivale a muestrear los 365 días del año sin reemplazo, por lo que el número de posibilidades es $365 \times 364 \times 363 \times \cdots \times (365-k+1)$ para $k \leq 365$. Por lo tanto, la probabilidad de que no haya coincidencias de cumpleaños en un grupo de $k$ personas es:

$$P(\text{sin coincidencia de cumpleaños}) = \frac{365 \times 364 \times \cdots \times (365-k+1)}{365^k}$$

y la probabilidad de que al menos haya una coincidencia de cumpleaños es:

$$P(\text{al menos 1 coincidencia de cumpleaños}) = 1 - \frac{365 \times 364 \times \cdots \times (365-k+1)}{365^k}$$

```python
import matplotlib.pyplot as plt

# Rango de valores para k (número de personas)
k_values = range(1, 100)

# Listas para las probabilidades
p_no_match = []
p_at_least_one = []

# Cálculo de las probabilidades
for k in k_values:
    prob_no = 1.0
    for i in range(k):
        prob_no *= (365 - i) / 365
    p_no_match.append(prob_no)
    p_at_least_one.append(1 - prob_no)

# Crear el gráfico
plt.figure()
plt.plot(k_values, p_no_match, label="P(no hay cumpleaños repetidos)")
plt.plot(k_values, p_at_least_one, label="P(al menos un cumpleaños repetido)")

# Ejes
plt.xlabel("Número de personas (k)")
plt.ylabel("Probabilidad")
plt.title("Problema del cumpleaños")

# Mostrar solo 0, 0.5 y 1 en el eje Y
plt.yticks([0, 0.5, 1])

# Línea discontinua fina en y = 0.5
plt.axhline(y=0.5, linestyle='--', linewidth=1)

# Leyenda
plt.legend()

plt.show()
```

El gráfico anterior muestra la probabilidad de al menos una coincidencia de cumpleaños en función de $k$. El primer valor de $k$ para el cual la probabilidad de una coincidencia excede 0.5 es $k = 23$. Por lo tanto, en un grupo de 23 personas, hay más del 50% de probabilidad de que haya al menos una coincidencia de cumpleaños. En $k = 57$, la probabilidad de una coincidencia ya excede el 99%.

El problema del cumpleaños es mucho más que un juego divertido de fiesta, y mucho más que una forma de construir intuición sobre coincidencias; también hay aplicaciones importantes en estadística y ciencias de la computación.

**Ejemplo 1.4.12 (El error de Leibniz).** Si lanzamos dos dados justos, ¿qué es más probable: una suma de 11 o una suma de 12?

**Solución:**

Etiqueta los dados A y B, y considera cada dado como un sub-experimento. Por la regla de la multiplicación, hay 36 posibles resultados para pares ordenados de la forma (valor de A, valor de B), y son igualmente probables por simetría. De estos, $(5, 6)$ y $(6, 5)$ son favorables a una suma de 11, mientras que solo $(6, 6)$ es favorable a una suma de 12. Por lo tanto, una suma de 11 es dos veces más probable que una suma de 12; la probabilidad es $\frac{1}{18}$ para el primero, y $\frac{1}{36}$ para el segundo.

Sin embargo, Leibniz argumentó erróneamente que una suma de 11 y una suma de 12 son igualmente probables, afirmando que cada una de estas sumas puede obtenerse de una sola manera. Aquí Leibniz estaba cometiendo el error de tratar los dos dados como objetos indistinguibles, viendo $(5, 6)$ y $(6, 5)$ como el mismo resultado.

¿Cuáles son los antídotos para el error de Leibniz? Primero, deberíamos etiquetar los objetos en cuestión en lugar de tratarlos como indistinguibles. Si Leibniz hubiera etiquetado sus dados A y B, o verde y naranja, o izquierdo y derecho, no habría cometido este error. Segundo, antes de usar el conteo para probabilidad, deberíamos preguntarnos si la definición ingenua aplica.

### 1.4.2 Ajuste por sobreconteo

En muchos problemas de conteo, no es fácil contar directamente cada posibilidad una y solo una vez. Sin embargo, si somos capaces de contar cada posibilidad exactamente $c$ veces para algún $c$, entonces podemos ajustar dividiendo por $c$. Por ejemplo, si hemos contado exactamente el doble de cada posibilidad, podemos dividir por 2 para obtener el conteo correcto. Llamamos a esto ajustar por sobreconteo.

Un coeficiente binomial cuenta el número de subconjuntos de cierto tamaño para un conjunto, como el número de formas de elegir un comité de tamaño $k$ de un conjunto de $n$ personas. Los conjuntos y subconjuntos son por definición no ordenados, por ejemplo, $\{3, 1, 4\} = \{4, 1, 3\}$, así que estamos contando el número de formas de elegir $k$ objetos de $n$, sin reemplazo y sin distinguir entre los diferentes órdenes en los que podrían elegirse.

**Definición 1.4.14 (Coeficiente binomial).** Para cualesquiera enteros no negativos $k$ y $n$, el coeficiente binomial $\binom{n}{k}$, leído como "$n$ elige $k$", es el número de subconjuntos de tamaño $k$ para un conjunto de tamaño $n$.

**Teorema 1.4.15 (Fórmula del coeficiente binomial).** Para $k \leq n$, tenemos:

$$\binom{n}{k} = \frac{n(n-1)\cdots(n-k+1)}{k!} = \frac{n!}{(n-k)!k!}$$

Para $k > n$, tenemos $\binom{n}{k} = 0$.

**Ejemplo 1.4.17 (Oficiales del club).** En un club con $n$ personas, hay $n(n-1)(n-2)$ formas de elegir un presidente, vicepresidente y tesorero, y hay:

$$\binom{n}{3} = \frac{n(n-1)(n-2)}{3!}$$
