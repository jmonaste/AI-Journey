## Probabilidad y conteo

La matematicas son la lógica de la certidumbre y la probabilidad es la lógica de la incertidumbre. la probabilidad es muy util en muchos campos porque nos ayuda a entender y a explicar las variaciones, separar señales del ruido y modelar fenomenos complejos. 

La probabilidad proporciona procedimientos para resolver problemas de manera fundamentada, pero también puede dar lugar a trampas y paradojas.incluso Gottfried Wilhelm von Leibniz y Sir Isaac Newton, las dos personas que descubrieron de forma independiente el cálculo en el siglo XVII, no fueron inmunes a cometer errores básicos en probabilidad.

un aspecto sorprendetne de la probabilidad es que muchas veces es posible estudiar un problema via simulacion. 

# 1.2 Espacios muestrales y el Mundo de las Piedras (Pebble World)

El marco matemático de la probabilidad se construye en torno a conjuntos. Imagina que se realiza un experimento cuyo resultado es uno entre un conjunto de posibles resultados. Antes de que se lleve a cabo el experimento, se desconoce cuál será el resultado; después, el resultado «cristaliza» en el resultado real.
